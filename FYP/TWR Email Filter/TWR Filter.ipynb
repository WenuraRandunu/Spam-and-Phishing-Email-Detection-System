```python
def TWR():

    start_time = time.time()
    
    emailA = emailAddress.get()
    domain = emailA.split("@")[1]
    subject = emailSubject.get()
    body = emailBody.get("1.0", tk.END)
    
    email = subject + "\n" + body


##########################################################################################################################################################
    
    # Email Address Checker Whitelist Domain
    def EACWD():
        # Read the CSV file
        df = pd.read_csv('White List/Domain White List.csv')
    
        # Check if the input domain is in the list of domains
        if domain in df['White_Listed_Domain_List'].values:
            return "Ham"
        else:
            return "Spam"

##########################################################################################################################################################
    
    # Email Address Checker Whitelist Email
    def EACWE():
        # Read the CSV file
        df = pd.read_csv('White List/Email White List.csv')
    
        # Check if the input email is in the list of email addresses
        if emailAddress in df['White_Listed_Email_List'].values:
            return "Ham"
        else:
            return "Spam"

##########################################################################################################################################################
    
    # Email Address Checker Blacklist Domain
    def EACBD():
        # Read the CSV file
        df = pd.read_csv('Black List/Domain Black List.csv')
    
        # Check if the input domain is in the list of domains
        if domain in df['Black_Listed_Domain_List'].values:
            return "Spam"
        else:
            return "Ham"

##########################################################################################################################################################
    
    # Email Address Checker Blacklist Email
    def EACBE():
        # Read the CSV file
        df = pd.read_csv('Black List/Email Black List.csv')
    
        # Check if the input email is in the list of email addresses
        if emailAddress in df['Black_Listed_Email_List'].values:
            return "Spam"
        else:
            return "Ham"

##########################################################################################################################################################

    # Email Subject Checking Filter
    def ESCF():
        # Read the CSV file
        df = pd.read_csv('Black List/Subject Black List.csv')
    
        # Check if the input domain is in the list of domains
        if subject in df['Black_Listed_Subject_List'].values:
            return "Spam"
        else:
            return "Ham"
           
##########################################################################################################################################################

    # NLP Classification Model
    def NLPCM():
        
        # Load spaCy English language model
        nlp = spacy.load("en_core_web_sm")
        
        # Function to preprocess the input paragraph
        def preprocess_paragraph(email):
            # Convert paragraph to lowercase
            email = email.lower()
            
            # Tokenize paragraph using spaCy
            doc = nlp(email)
            
            # Remove stop words
            tokens_without_stopwords = [token.text for token in doc if token.text not in STOP_WORDS]
            
            return tokens_without_stopwords
        
        # Function to count word occurrences
        def count_word_occurrences(tokens):
            word_counts = Counter(tokens)
            return word_counts
        
        # Function to get word values from Word_List.csv
        def get_word_values(word_counts, word_list_path):
            word_list = pd.read_csv(word_list_path)
            word_values = {}
            total_value_paragraph = 0  # Total value of words in the paragraph
            total_value_matching = 0   # Total value of matching words
            
            for index, row in word_list.iterrows():
                word = str(row["Words"]).lower()  # Convert to string to handle NaN values
                if word != "nan" and word in word_counts:  # Check for NaN values
                    value = word_counts[word]
                    word_values[word] = value
                    total_value_matching += value
                    
            # Calculate total value of words in the paragraph
            total_value_paragraph = sum(word_counts.values())
        
            if total_value_paragraph == 0:
                total_value_paragraph = 1
        
            spam_ratio = total_value_matching/total_value_paragraph
            
            return word_values, total_value_paragraph, total_value_matching
    
        
        # Preprocess paragraph
        tokens = preprocess_paragraph(email)
        
        # Count word occurrences
        word_counts = count_word_occurrences(tokens)
        
        # Word_List.csv path
        word_list_path = "Word List/Word_List.csv"
        
        # Get word values from Word_List.csv
        word_values, total_value_paragraph, total_value_matching = get_word_values(word_counts, word_list_path)
        
        spam_ratio = total_value_matching / total_value_paragraph
    
        
        if spam_ratio > 0.20:
            return "Spam"
        else:
            return "Ham" 

##########################################################################################################################################################

    # Load dataset
    df = pd.read_csv("Data Set/TWR.csv")
    
    # Assuming your dataset has 'text' column for email content and 'label' column for spam/ham labels
    X = df['Message']
    y = df['Category']
    
    # Map labels to numerical values
    label_mapping = {'spam': 1, 'ham': 0}
    y = y.map(label_mapping)
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    
##########################################################################################################################################################
    
    # Multinomial Naive Bayes Classifiction Model
    
    def MNBCM():
        
        # Create a pipeline with TfidfVectorizer and Multinomial Naive Bayes
        pipeline = Pipeline([
            ('tfidf', TfidfVectorizer(
                lowercase=True,
                stop_words='english',
                max_df=0.8,
                min_df=0.01,
                ngram_range=(1, 2)  # Include unigrams and bigrams
            )),
            ('nb', MultinomialNB())
        ])
    
        # Define hyperparameter grid for GridSearchCV
        param_grid = {
            'tfidf__max_features': [1000, 5000, None],
            'nb__alpha': [0.1, 1, 10]
        }
    
        # Use grid search to find optimal hyperparameters
        grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='accuracy')
        grid_search.fit(X_train, y_train)
    
        # Get the best model from grid search
        best_model = grid_search.best_estimator_
    
        # Make predictions on the test set
        predictions_proba = best_model.predict_proba(X_test)[:, 1]  # Get the probability of being spam
    
        # Adjust the threshold (experiment with different values)
        threshold = 0.95
        predictions = (predictions_proba > threshold).astype(int)
    
        # Evaluate the model
        # accuracy = metrics.accuracy_score(y_test, predictions)
        # print(f"Accuracy: {accuracy}")
    
        # Function to check if an email is spam or not
        def is_spam(email_text):
            email_text = BeautifulSoup(email_text, 'html.parser').get_text()
            email_tfidf = best_model.named_steps['tfidf'].transform([email_text])
            prediction_proba = best_model.named_steps['nb'].predict_proba(email_tfidf)[:, 1]
            return prediction_proba[0]
    
        # Example usage
        new_email = email
    
        spam_probability = is_spam(new_email)
    
        if spam_probability > threshold:
            return "Spam"
        else:
            if (1-spam_probability) < threshold:
                return "Spam"
            else:
                return "Ham"
    
##########################################################################################################################################################
 
    # Logistic Regression Classfication Model
    
    def LRCM():

        # Create a pipeline with TfidfVectorizer and Logistic Regression
        pipeline = Pipeline([
            ('tfidf', TfidfVectorizer(
                lowercase=True,
                stop_words='english',
                max_df=0.8,
                min_df=0.01,
                ngram_range=(1, 2)  # Include unigrams and bigrams
            )),
            ('logreg', LogisticRegression(solver='liblinear', max_iter=1000))
        ])
    
        # Define hyperparameter grid for GridSearchCV
        param_grid = {
            'tfidf__max_features': [1000, 5000, None],
            'logreg__C': [0.1, 1, 10],
            'logreg__class_weight': ['balanced', None]
        }
    
        # Use grid search to find optimal hyperparameters
        grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='accuracy')
        grid_search.fit(X_train, y_train)
    
        # Get the best model from grid search
        best_model = grid_search.best_estimator_
    
        # Make predictions on the test set
        predictions_proba = best_model.predict_proba(X_test)[:, 1]  # Get the probability of being spam
    
        # Adjust the threshold (experiment with different values)
        threshold = 0.95
        predictions = (predictions_proba > threshold).astype(int)
    
        # Evaluate the model
        # accuracy = metrics.accuracy_score(y_test, predictions)
        # print(f"Accuracy: {accuracy}")
    
        # Function to check if an email is spam or not
        def is_spam(email_text):
            email_text = BeautifulSoup(email_text, 'html.parser').get_text()
            email_tfidf = best_model.named_steps['tfidf'].transform([email_text])
            prediction_proba = best_model.named_steps['logreg'].predict_proba(email_tfidf)[:, 1]
            return prediction_proba[0]
    
        # Example usage
        new_email = email
    
        spam_probability = is_spam(new_email)
    
        if spam_probability > threshold:
            return "Spam" 
        else:
            if (1-spam_probability) < threshold:
                return "Spam"
            else:
                return "Ham"
    

##########################################################################################################################################################
    # K-Nearest Neighbors CLassification Model
    
    def KNNCM():
    
        # Create a pipeline with TfidfVectorizer and K-Nearest Neighbors
        pipeline = Pipeline([
            ('tfidf', TfidfVectorizer(
                lowercase=True,
                stop_words='english',
                max_df=0.8,
                min_df=0.01,
                ngram_range=(1, 2)  # Include unigrams and bigrams
            )),
            ('knn', KNeighborsClassifier())
        ])
    
        # Define hyperparameter grid for GridSearchCV
        param_grid = {
            'tfidf__max_features': [1000, 5000, None],
            'knn__n_neighbors': [3, 5, 7],
            'knn__weights': ['uniform', 'distance']
        }
    
        # Use grid search to find optimal hyperparameters
        grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='accuracy')
        grid_search.fit(X_train, y_train)
    
        # Get the best model from grid search
        best_model = grid_search.best_estimator_
    
        # Make predictions on the test set
        predictions_proba = best_model.predict_proba(X_test)[:, 1]  # Get the probability of being spam
    
        # Adjust the threshold (experiment with different values)
        threshold = 0.95
        predictions = (predictions_proba > threshold).astype(int)
    
        # Evaluate the model
        # accuracy = metrics.accuracy_score(y_test, predictions)
        # print(f"Accuracy: {accuracy}")
    
        # Function to check if an email is spam or not
        def is_spam(email_text):
            email_text = BeautifulSoup(email_text, 'html.parser').get_text()
            email_tfidf = best_model.named_steps['tfidf'].transform([email_text])
            prediction_proba = best_model.named_steps['knn'].predict_proba(email_tfidf)[:, 1]
            return prediction_proba[0]
    
        # Example usage
        new_email = email
    
        spam_probability = is_spam(new_email)
    
        if spam_probability > threshold:
            return "Spam" 
        else:
            if (1-spam_probability) < threshold:
                return "Spam"
            else:
                return "Ham"


##########################################################################################################################################################
  
    # Support Vector Machine Classification Model
    
    def SVMCM():
    
        # Create a pipeline with TfidfVectorizer and Linear SVM
        pipeline = Pipeline([
            ('tfidf', TfidfVectorizer(
                lowercase=True,
                stop_words='english',
                max_df=0.8,
                min_df=0.01,
                ngram_range=(1, 2)  # Include unigrams and bigrams
            )),
            ('svm', SVC(kernel='linear', probability=True))
        ])
    
        # Define hyperparameter grid for GridSearchCV
        param_grid = {
            'tfidf__max_features': [1000, 5000, None],
            'svm__C': [0.1, 1, 10],
            'svm__class_weight': ['balanced', None]
        }
    
        # Use grid search to find optimal hyperparameters
        grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='accuracy')
        grid_search.fit(X_train, y_train)
    
        # Get the best model from grid search
        best_model = grid_search.best_estimator_
    
        # Make predictions on the test set
        predictions_proba = best_model.predict_proba(X_test)[:, 1]  # Get the probability of being spam
    
        # Adjust the threshold (experiment with different values)
        threshold = 0.95
        predictions = (predictions_proba > threshold).astype(int)
    
        # Evaluate the model
        # accuracy = metrics.accuracy_score(y_test, predictions)
        # print(f"Accuracy: {accuracy}")
    
        # Function to check if an email is spam or not
        def is_spam(email_text):
            email_text = BeautifulSoup(email_text, 'html.parser').get_text()
            email_tfidf = best_model.named_steps['tfidf'].transform([email_text])
            prediction_proba = best_model.named_steps['svm'].predict_proba(email_tfidf)[:, 1]
            return prediction_proba[0]
    
        # Example usage
        new_email = email
    
        spam_probability = is_spam(new_email)
    
        if spam_probability > threshold:
            return "Spam" 
        else:
            if (1-spam_probability) < threshold:
                return "Spam"
            else:
                return "Ham"

##########################################################################################################################################################
    # Gradient Boosting Machines Classification Model
    
    def GBMCM():
    
        # Create a pipeline with TfidfVectorizer and Gradient Boosting Classifier
        pipeline = Pipeline([
            ('tfidf', TfidfVectorizer(
                lowercase=True,
                stop_words='english',
                max_df=0.8,
                min_df=0.01,
                ngram_range=(1, 2)  # Include unigrams and bigrams
            )),
            ('gb', GradientBoostingClassifier())
        ])
    
        # Define hyperparameter grid for GridSearchCV
        param_grid = {
            'tfidf__max_features': [1000, 5000, None],
            'gb__n_estimators': [100, 200],
            'gb__learning_rate': [0.1, 0.05],
            'gb__max_depth': [3, 4]
        }
    
        # Use grid search to find optimal hyperparameters
        grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='accuracy')
        grid_search.fit(X_train, y_train)
    
        # Get the best model from grid search
        best_model = grid_search.best_estimator_
    
        # Make predictions on the test set
        predictions_proba = best_model.predict_proba(X_test)[:, 1]  # Get the probability of being spam
    
        # Adjust the threshold (experiment with different values)
        threshold = 0.95
        predictions = (predictions_proba > threshold).astype(int)
    
        # Evaluate the model
        # accuracy = metrics.accuracy_score(y_test, predictions)
        # print(f"Accuracy: {accuracy}")
    
        # Function to check if an email is spam or not
        def is_spam(email_text):
            email_text = BeautifulSoup(email_text, 'html.parser').get_text()
            email_tfidf = best_model.named_steps['tfidf'].transform([email_text])
            prediction_proba = best_model.named_steps['gb'].predict_proba(email_tfidf)[:, 1]
            return prediction_proba[0]
    
        # Example usage
        new_email = email
    
        spam_probability = is_spam(new_email)
    
        if spam_probability > threshold:
            return "Spam" 
        else:
            if (1-spam_probability) < threshold:
                return "Spam"
            else:
                return "Ham"

##########################################################################################################################################################
    
    if EACWD() == "Ham":
        text_area_2.delete('1.0', END)
        text_area_2.insert(tk.INSERT, "Ham EACWD")
    else:
        if EACWE() == "Ham":
            text_area_2.delete('1.0', END)
            text_area_2.insert(tk.INSERT, "Ham EACWE")
        else:
            if EACBD() == "Spam":
                text_area_2.delete('1.0', END)
                text_area_2.insert(tk.INSERT, "Spam EACBD")
                
                with open('Detected Spam/Detected Spam Email List.csv', 'a', newline='') as f:
                    pd.DataFrame([[emailA]], columns=['Spam_Email_Address']).to_csv(f, header=f.tell() == 0, index=False)
                    f.close()
                
                with open('Detected Spam/Detected Spam Email Subject List.csv', 'a', newline='') as f:
                    pd.DataFrame([[subject]], columns=['Spam_Subjects']).to_csv(f, header=f.tell() == 0, index=False)
                    f.close()
                    
            else:
                if EACBE() == "Spam":
                    text_area_2.delete('1.0', END)
                    text_area_2.insert(tk.INSERT, "Spam EACBE")
                    
                    with open('Detected Spam/Detected Spam Email List.csv', 'a', newline='') as f:
                        pd.DataFrame([[emailA]], columns=['Spam_Email_Address']).to_csv(f, header=f.tell() == 0, index=False)
                        f.close()
                    
                    with open('Detected Spam/Detected Spam Email Subject List.csv', 'a', newline='') as f:
                        pd.DataFrame([[subject]], columns=['Spam_Subjects']).to_csv(f, header=f.tell() == 0, index=False)
                        f.close()
                        
                else:
                    if ESCF() == "Spam":
                        text_area_2.delete('1.0', END)
                        text_area_2.insert(tk.INSERT, "Spam ESCF")
                        
                        with open('Detected Spam/Detected Spam Email List.csv', 'a', newline='') as f:
                            pd.DataFrame([[emailA]], columns=['Spam_Email_Address']).to_csv(f, header=f.tell() == 0, index=False)
                            f.close()
                        
                        with open('Detected Spam/Detected Spam Email Subject List.csv', 'a', newline='') as f:
                            pd.DataFrame([[subject]], columns=['Spam_Subjects']).to_csv(f, header=f.tell() == 0, index=False)
                            f.close()
                            
                    else:
                        if NLPCM() == "Spam":
                            text_area_2.delete('1.0', END)
                            text_area_2.insert(tk.INSERT, "Spam NLPCM")
                            
                            with open('Detected Spam/Detected Spam Email List.csv', 'a', newline='') as f:
                                pd.DataFrame([[emailA]], columns=['Spam_Email_Address']).to_csv(f, header=f.tell() == 0, index=False)
                                f.close()
                            
                            with open('Detected Spam/Detected Spam Email Subject List.csv', 'a', newline='') as f:
                                pd.DataFrame([[subject]], columns=['Spam_Subjects']).to_csv(f, header=f.tell() == 0, index=False)
                                f.close()
                                
                        else:
                            if MNBCM() == "Spam":
                                text_area_2.delete('1.0', END)
                                text_area_2.insert(tk.INSERT, "Spam MNBCM")
                                
                                with open('Detected Spam/Detected Spam Email List.csv', 'a', newline='') as f:
                                    pd.DataFrame([[emailA]], columns=['Spam_Email_Address']).to_csv(f, header=f.tell() == 0, index=False)
                                    f.close()
                                
                                with open('Detected Spam/Detected Spam Email Subject List.csv', 'a', newline='') as f:
                                    pd.DataFrame([[subject]], columns=['Spam_Subjects']).to_csv(f, header=f.tell() == 0, index=False)
                                    f.close()
                                
                            else:
                                if LRCM() == "Spam":
                                    text_area_2.delete('1.0', END)
                                    text_area_2.insert(tk.INSERT, "Spam LRCM")
                                    
                                    with open('Detected Spam/Detected Spam Email List.csv', 'a', newline='') as f:
                                        pd.DataFrame([[emailA]], columns=['Spam_Email_Address']).to_csv(f, header=f.tell() == 0, index=False)
                                        f.close()
                                    
                                    with open('Detected Spam/Detected Spam Email Subject List.csv', 'a', newline='') as f:
                                        pd.DataFrame([[subject]], columns=['Spam_Subjects']).to_csv(f, header=f.tell() == 0, index=False)
                                        f.close()
                                        
                                else:
                                    if KNNCM() == "Spam":
                                        text_area_2.delete('1.0', END)
                                        text_area_2.insert(tk.INSERT, "Spam KNNCM")
                                        
                                        with open('Detected Spam/Detected Spam Email List.csv', 'a', newline='') as f:
                                            pd.DataFrame([[emailA]], columns=['Spam_Email_Address']).to_csv(f, header=f.tell() == 0, index=False)
                                            f.close()
                                        
                                        with open('Detected Spam/Detected Spam Email Subject List.csv', 'a', newline='') as f:
                                            pd.DataFrame([[subject]], columns=['Spam_Subjects']).to_csv(f, header=f.tell() == 0, index=False)
                                            f.close()
                                            
                                    else:
                                        if SVMCM() == "Spam":
                                            text_area_2.delete('1.0', END)
                                            text_area_2.insert(tk.INSERT, "Spam SVMCM")
                                            
                                            with open('Detected Spam/Detected Spam Email List.csv', 'a', newline='') as f:
                                                pd.DataFrame([[emailA]], columns=['Spam_Email_Address']).to_csv(f, header=f.tell() == 0, index=False)
                                                f.close()
                                            
                                            with open('Detected Spam/Detected Spam Email Subject List.csv', 'a', newline='') as f:
                                                pd.DataFrame([[subject]], columns=['Spam_Subjects']).to_csv(f, header=f.tell() == 0, index=False)
                                                f.close()
                                                
                                        else:
                                            if GBMCM() == "Spam":
                                                text_area_2.delete('1.0', END)
                                                text_area_2.insert(tk.INSERT, "Spam GBMCM")
                                                
                                                with open('Detected Spam/Detected Spam Email List.csv', 'a', newline='') as f:
                                                    pd.DataFrame([[emailA]], columns=['Spam_Email_Address']).to_csv(f, header=f.tell() == 0, index=False)
                                                    f.close()
                                                
                                                with open('Detected Spam/Detected Spam Email Subject List.csv', 'a', newline='') as f:
                                                    pd.DataFrame([[subject]], columns=['Spam_Subjects']).to_csv(f, header=f.tell() == 0, index=False)
                                                    f.close()
                                                    
                                            else:
                                                text_area_2.delete('1.0', END)
                                                text_area_2.insert(tk.INSERT, "Ham")
                                        
                            
    end_time = time.time()

    processTime = end_time - start_time

    text_area_2.insert(tk.INSERT, "\n" + str(processTime))

    
        
```
